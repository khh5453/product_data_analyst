```python
import pandas as pd
import numpy as np
```

# Step 1

### 1. cars.csv와 brand.csv의 데이터 합치기
[힌트] cars.csv의 title 변수로부터 브랜드 데이터를 추출할 수 있습니다.

```python
cars = pd.read_csv('cars.csv')
cars.info()
cars.head()
```

```python
brand = pd.read_csv('brand.csv')
brand.info()
brand.head()
```

```python
### cars 데이터프레임의 title 컬럼에서 브랜드명만 추출하기
cars['brand'] = cars['title'].str.split().apply(lambda x : x[0])
cars['title'] = cars['title'].str.split().apply(lambda x : x[1])
cars
```

```python
### brand 데이터프레임에서 title 컬럼 값을 대문자로 바꿔주기
brand['title'] = brand['title'].str.upper()
brand.head()
```

```python
### cars, brand 데이터프레임을 병합하기
df_cars = cars.merge(brand, left_on='brand', right_on='title', how='left')
df_cars.drop('title_y', axis=1, inplace=True)
df_cars.rename({'title_x' : 'title'}, axis=1, inplace=True)
df_cars.info()
```

### 2. 카테고리형 변수를 숫자 형태로 변환하기
[힌트] Engine, Emission Class 변수에 주목하세요!

```python
df_cars.head()
```

```python
### Engine 컬럼에서 각 값 마지막에 있는 'L'을 빼고 수치형 타입으로 바꿔주기
df_cars['Engine'] = df_cars['Engine'].str.strip('L').astype('float')
df_cars
```

```python
### Emission Class 컬럼에서 'Euro' 라는 값을 빼고 숫자만 남긴 후 수치형 타입으로 바꿔주기
df_cars['Emission Class'].value_counts()
df_cars['Emission Class'] = df_cars['Emission Class'].str.strip('Euro ').astype('float')
df_cars.info()
```

```python
df_cars.select_dtypes(exclude='number').nunique()
```

### 3. 결측치 처리하기
[힌트] Service history 변수의 결측치는 'Unknown'으로 채우고, 결측치 변수가 일정 개수 이상 포함된 데이터는 제거합니다.

```python
### Service history 변수의 결측치는 'Unknown'으로 채우기
df_cars['Service history'] = df_cars['Service history'].fillna('Unknown')
df_cars['Service history'].isna().sum()
```

```python
### 각 행마다 결측치 개수를 세어서 결측치 개수가 4개 이상인 행의 인덱스 추출
temp = df_cars.isnull().sum(axis=1)
idx = [i for i in range(len(temp.index)) if temp[i] > 3]

### 해당 인덱스의 값 확인
df_cars.iloc[idx,:]

### 해당 행 제거하기
df_cars = df_cars.drop(idx, axis=0).reset_index(drop=True)
df_cars
```

```python
df_cars.isna().mean()
### 각 컬럼별 결측치 비율을 보았을 때, 50% 이하이므로 결측치 대체하기로 결정

df_cars.describe()
### 각 컬럼의 describe 값을 보았을 때, 이상치도 없으므로 평균으로 대체하기로 결정
```

```python
### 각 컬럼의 결측치를 평균으로 대체하기
po_mean = round(df_cars['Previous Owners'].mean())
df_cars['Previous Owners'] = df_cars['Previous Owners'].fillna(po_mean)

eg_mean = round(df_cars['Engine'].mean(), 1)
df_cars['Engine'] = df_cars['Engine'].fillna(eg_mean)

door_mean = round(df_cars['Doors'].mean())
df_cars['Doors'] = df_cars['Doors'].fillna(door_mean)

sheet_mean = round(df_cars['Seats'].mean())
df_cars['Seats'] = df_cars['Seats'].fillna(sheet_mean)

ec_mean = round(df_cars['Emission Class'].mean())
df_cars['Emission Class'] = df_cars['Emission Class'].fillna(ec_mean)

df_cars.isna().sum()
```

### 범주형 변수를 수치형 변수로 바꾸기

```python
df_cars.head()
```

```python
df_cars.select_dtypes(exclude='number').nunique()
```

```python
### title 컬럼은 범주 종류가 너무 많고 값이 퍼져있으므로 모델 학습에 도움이 되지 않을 것으로 판단하여 제거하기로 결정
df_cars = df_cars.drop('title', axis=1)
df_cars
```

```python
### brand 컬럼의 각 고유값의 개수가 50 이하인 경우 others 로 변경하기로 결정
temp_index = df_cars['brand'].value_counts().sort_values().index[:29]
replace_dict = {i:'others' for i in temp_index}
df_cars['brand'] = df_cars['brand'].replace(replace_dict)
df_cars['brand'].value_counts()
```

```python
### 원-핫 인코딩 사용하기
cols = df_cars.select_dtypes(exclude='number').columns
cars_encoded = pd.get_dummies(df_cars, columns=cols, drop_first=True, dtype=int)
cars_encoded
```

### 4. Scaler 활용해보기

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
cars_scaled = ss.fit_transform(cars_encoded)
cars_scaled = pd.DataFrame(cars_scaled, columns=cars_encoded.columns)
cars_scaled
```

### 5. 기존 데이터의 정보 70% 이상을 가지는 수준에서 최소한의 주성분 추출 (PCA)

```python
from sklearn.decomposition import PCA
pca = PCA()
a = pca.fit_transform(cars_scaled)
b = pd.DataFrame(a, columns=[f'pc{i+1}' for i in range(a.shape[1])])
b.head()
```

```python
result = pd.DataFrame({'설명가능한 분산 비율(고윳값)':pca.explained_variance_,
                       '기여율':pca.explained_variance_ratio_},
                      index = np.array([f'pc{i+1}' for i in range(a.shape[1])]))
result['누적기여율'] = result['기여율'].cumsum()
(result['누적기여율'] <= 0.7).sum()

### 기존 데이터의 정보 70% 이상을 가지는 최소한의 주성분 개수를 21개로 판단함
# -> (result['누적기여율'] <= 0.7).sum() 해당 식에서는 0.7미만의 값만 포함하므로 +1 해주었음

### 주성분 개수 구하는 다른 방법
# pca2 = PCA(n_components=0.7)
# pca2.fit_transform(cars_scaled).shape
```

```python
pca = PCA(n_components=21)
np_pca = pca.fit_transform(cars_scaled)
df_pca = pd.DataFrame(np_pca, columns=[f'pc{i+1}' for i in range(np_pca.shape[1])])
df_pca
```
